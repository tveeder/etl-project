```python
import pandas as pd
#import numpy as np
#import datetime
```

```python

coup_df = pd.read_csv('Resources/coup-and-commodities.csv')
coup_df.head()
```

```python
coup_df = coup_df[['date', 'country', 'couprisk']].copy()
coup_df.head()
```

```python

economic_df = pd.read_csv('Resources/economic-crisis-africa.csv')
economic_df.head()
```

```python

economic_df = economic_df[['country', 'year','banking_crisis']].copy()
economic_df.head()
```

```python
coup_df.dtypes
```

```python
# extract just the year to merger with other dataset

coup_df['date'] = pd.to_datetime(coup_df['date'])
coup_df['year']= coup_df['date'].apply(lambda x: x.year)


```

```python
coup_df.head()
```

```python
merged = pd.merge(coup_df,economic_df, on=["country",'year'],how="inner")
```
